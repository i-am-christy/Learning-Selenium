```python
#import beautiful soup library
from bs4 import BeautifulSoup
#import the urlopen to open urls
from urllib.request import urlopen
```

```python
#specifying the link to the website to scrap from
url = 'http://www.hubertiming.com/results/2017GPTR10K'
#opeing the link
html = urlopen(url)
```

```python
#getting the html object of the website
#the soup object allows us to access several parts of the html code of the page,like the title, 
# headeers, etc
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
#get the title of the page
title = soup.title
print(title)
```

```python
#find all the table, specifically the table that contains the data that we are looking for
table = soup.find_all('table')[1]
print(table)
```

```python
#getting the rows in the table in the website
rows = table.find_all('tr')
table_data = []

#looping through the row data to obtain the table data
for row in rows:
    row_data = row.find_all('td')
    individual_data = [data.text.strip() for data in row_data]
    table_data.append(individual_data)
```

```python
#checking the data and removing the first empty list from the data
table_data.pop(0)
table_data
```

```python
#get the header from the table
headers = table.find_all('th')
table_header = [header.text.strip() for header in headers]
```

```python
table_header
```

```python
#covert the data to dataframe
import pandas as pd
df = pd.DataFrame(table_data, columns=table_header)
```
