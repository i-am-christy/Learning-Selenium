```python
#creating the spider object
from typing import Iterable


from scrapy.http import Request


class MySpider(scrapy.Spider):
    name = "my_spider"

    #create the start request 
    def start_requests(self):
        url = r'https://www.leolist.cc/motors/cars-trucks-rv/greater-toronto'
        yield scrapy.Request(url= url, callback= self.parse_first)


    def parse_first(self, reponse):
        #narrow in on each list iteam
        div_list = response.css('div.lst lst-item')
        #narrow down to the links 
        ad_links = div_list.xpath('./a/@href')
        #extract the links as a list of strings
        links_to_follow = ad_links.extract()
        for url in links_to_follow:
            yield reponse.follow(url= url, callback= self.parse_image)

    #parser to get the image link
    def parse_image(self, response):
        #direct to item title
        item_name = response.css('div.value::text').strip
        #direct to the image gallery div and image link
        gallery_div = response.css('div.simple-gallery').xpath('./a/@href')
        #direct to the image and extract into a list
        img_list = gallery_div.extract_first().strip()
        #img_list_ext = [img.strip() for img in img_list]
        #img_iteam_dict[ ]

#create a list
img_list = list()

#create the crawler object
process = CrawlerProcess()

#call  your spider
process.crawl(MySpider)

#start crawling
process.start()
```
